```python
import pandas as pd
import re
from datetime import date
```

```python
data = pd.read_csv("C:\\Users\\elias\\Documents\\GitHub\\Maze\project_4\\pr_4\\main_task_new.csv")
```

```python
data.sample(5)
```

```python
# чистим колонку разновидностей Кухни
data['Cuisine Style'] = data['Cuisine Style'].str.replace('[','')
data['Cuisine Style'] = data['Cuisine Style'].str.replace(']','')
data['Cuisine Style'] = data['Cuisine Style'].str.replace("'",'')

# и выводим в список
data['Cuisine Style'] = data['Cuisine Style'].str.split(', ')
```

```python
#Выводим dummy признак видов кухни 
s = data['Cuisine Style']
cuisine = pd.get_dummies(s.apply(pd.Series).stack()).sum(level=0)

#... и объединяем
data = pd.concat([data,cuisine], axis = 1)
```

```python
#Выводим dummy признак City
data = pd.get_dummies(data, columns = ['City'], dummy_na = True)
```

```python
#чистим колонку Restaraunt ID и переводим в числовой тип
data['Restaurant_id'] = data['Restaurant_id'].str.replace('id_','')
data['Restaurant_id'] = data['Restaurant_id'].astype(int)

#чистим колонку ID_TA и переводим в числовой тип
data['ID_TA'] = data['ID_TA'].str.replace('d','')
data['ID_TA'] = data['ID_TA'].astype(int)
```

```python
#очистка колонки Price Range
data['Price Range'] = data['Price Range'].str.replace('$','1')
data['Price Range'] = data['Price Range'].str.replace('11 - 111','3')
data['Price Range'] = data['Price Range'].str.replace('1111','2')

# заполнение пропусков и конвертация в числовой тип
data['Price Range'] = data['Price Range'].fillna('0')
data['Price Range'] = data['Price Range'].astype(int)
```

```python
#заполняем все пропуски
data = data.fillna(0)
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split

# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = data.drop(['Cuisine Style','Rating','Reviews', 'URL_TA'], axis = 1)
y = data['Rating']
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)
```

```python
# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)
```

```python
# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
